### 主要是抓取表特版圖片，但由於光是抓取圖片跟標題，後續要再查找該文章會比較麻煩
### 於是用這個方式做個整理，將文章以及相關資訊額外放置於excel
### 單純供學術練習使用

```python
# 表特
import requests
import shutil
import os
import pandas as pd
import time
from bs4 import BeautifulSoup

start = time.time()
print("start", start)

url = "https://www.ptt.cc/bbs/Beauty/index.html" # ptt表特版 

# 如果沒有資料夾 則創建資料夾 後續放置圖片以及excel
img_path = "./image"
excel_path = "./image_url"
if not os.path.isdir(img_path):
    os.mkdir(img_path)
if not os.path.isdir(excel_path):
    os.mkdir(excel_path)

# 放置url 後續產成excel 
excel_list = []

# 是否已滿 18歲驗證
def over18(url):
    response = requests.get(url)
    response = requests.get(url, cookies = {'over18': '1'})
    return response

# 抓 href
def get_all_href(url):
    r = over18(url)
    soup = BeautifulSoup(r.text, "lxml")
    
    results = soup.select("div.title")
    for item in results:
        a_item = item.select_one("a")
        if a_item:
            a_item_href = str(a_item.get('href'))
            href_id_list = a_item_href.split('.')
            href_id = href_id_list[1] + '_' + href_id_list[2] + '_' + href_id_list[3]
            get_article_content(article_url = 'https://www.ptt.cc' + a_item_href, img_prefix = href_id)
            
    print('------- next page -----')

# 抓取文章內容
def get_article_content(article_url, img_prefix):
    print('img_prefix:', img_prefix)
    r = over18(article_url)
    soup = BeautifulSoup(r.text, "lxml")
    results = soup.select('span.article-meta-value')
    if results:
        # 作者、看板、標題、時間、url        
        excel_list.append([results[0].text, results[1].text, results[2].text, results[3].text, article_url])        
        image_count = 0
        imgs = soup.findAll('a')
        for img in imgs:
            if '.jpg' in img['href']:
                download_img_from_article(img_url=img['href'], img_name = img_prefix + str(image_count))
                image_count += 1    

# 下載圖片
def download_img_from_article(img_url, img_name):
    r = requests.get(img_url, stream=True)
    file_name = str(img_name)
    print( 'save img to  ./image/'+ file_name + '.jpg')
    try:
        with open('./image/' + file_name + '.jpg', 'wb+') as out_file:
            shutil.copyfileobj(r.raw, out_file)
    except:
        print('can not save img', img_url)
    
# 判斷抓幾頁跑迴圈
# 先寫3頁 如果之後要提供使用再調整成使用者輸入
for page in range(0, 3):
    try:
        r = over18(url)
        soup = BeautifulSoup(r.text, "lxml")
        btn = soup.select('div.btn-group > a')
        if btn:
            next_page_url = 'https://www.ptt.cc' + btn[3]['href']
            url = next_page_url
            print('page count:', page)
            print('this page:', url)
            get_all_href(url = url)
    except Exception as e:
        print('An Error occurred: ', e)
        
df = pd.DataFrame(excel_list, columns=['作者', '看板', '標題', '時間', 'url'])
df.to_excel("./image_url/output.xlsx", sheet_name='表特版url整理')
print("done")

end = time.time()
print("end", end)
print(end - start)
```
